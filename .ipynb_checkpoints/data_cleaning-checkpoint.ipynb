# 2. Gold Layer

```python
import pandas as pd
from datetime import datetime

# Assume que dfS jÃ¡ existe
dfG = pd.DataFrame(columns=dfS.columns.tolist() + ['created_date', 'last_updated_date'])

# Data atual
now = datetime.now()

# FunÃ§Ã£o para inserir ou atualizar
def update_or_create(dfG, dfS):
    dfG = dfG.copy()

    for _, row in dfS.iterrows():
        code = row['id']

        # Verifica se o cÃ³digo jÃ¡ existe em dfG
        existing = dfG[dfG['id'] == code]

        if not existing.empty:
            # Existe -> cria novo registro copiando todas as colunas
            new_row = row.to_dict()
            new_row['created_date'] = now
            new_row['last_updated_date'] = now
            dfG = pd.concat([dfG, pd.DataFrame([new_row])], ignore_index=True)
        else:
            # NÃ£o existe -> cria novo registro copiando todas as colunas
            new_row = row.to_dict()
            new_row['created_date'] = now
            new_row['last_updated_date'] = now
            dfG = pd.concat([dfG, pd.DataFrame([new_row])], ignore_index=True)
    
    return dfG

# Atualiza dfG baseado em dfS
dfG = update_or_create(dfG, dfS)
dfG.to_csv('data_lakehouse/gold/starG.csv', index=False)  
dfG
```
